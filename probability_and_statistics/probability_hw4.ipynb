### Задача 1

О непрерывной равномерно распределенной случайной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

**Решение**

Математическое ожидание и дисперсия равномерного распределения:

$$M(B) = \dfrac{a + b}{2}, \text{ } D(B) = \dfrac{(b - a)^2}{12}.$$

Имеем:

**D(B) = 0.2**

**a = 0.5**

Чтобы найти правую границу(b), подставим в формулу дисперсии имеющиеся значения, и раскроем скобки по формуле разности квадратов. Получим квадратное уравнение:

$$b^2+2ab-a^2 = D(B) * 12$$

$$b^2+b-0.25 = 2.4$$

$$b^2+b-2.65 = 0$$

$$D = 11.6$$


Теперь с помощью дискриминанта найдем корни:

$$b_1 \approx -2.2029$$

$$b_2 \approx  1.2029$$

Так как мы ищем **правую границу** искомый корень должен быть **больше 0.5**

Соответственно, правая граница этого распределения приблизительно равна **1.2029**

Теперь имея обе границы можем найти мат. ожидание:

$$M(B) \approx \dfrac{0.5 + 1.2029}{2}$$

$$M(B) \approx 0.85145$$

### Задача 2

На сколько сигм (средних квадратических отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 и D(X) = 25?

**Решение**

Здесь мы имеем дело с **нормальным распределением**

Для начала найдем сигму используя формулу:

$$D(X) = \sigma^2$$

$$\sigma = \sqrt {D(X)}$$

Получаем:

$$\sigma = 5$$

Чтобы ответить на вопрос из задания, найдем разницу между имеющимся ростом и мат. ожиданием и разделим эту разницу на ранее найденую нами сигму.

$$\frac {190 - 178}{5} = 2.4$$

**Ответ:**

На **2.4 сигмы** отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 и D(X) = 25.

### Задача 3

Рост взрослого населения города X имеет нормальное распределение, причем, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. Без использования сторонних библиотек (numpy, scipy, pandas и пр.) посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
1. больше 182 см?
2. больше 190 см?
3. от 166 см до 190 см?
4. от 166 см до 182 см?
5. от 158 см до 190 см?
6. не выше 150 см или не ниже 190 см?
7. не выше 150 см или не ниже 198 см?
8. ниже 166 см?

Подсказка: всё необходимое для решения этого задания было в уроке 4.

**Решение**

Имеем:

$$a = 174$$

$$\sigma = 8$$

Сначала я собирался использовать функицю Лапласа, ибо в интернете везде я находил ее как самый часто применяемый способ.

Но затем внимательно посмотрев на интервалы, вероятность попадания в которые нужно найти, я понял, что все они отличаются от мат ожидания на целое число сигм.

Исходя из этого, будем находить вероятности по **правилам сигм**:

* Интервал от $a - \sigma$ до $a + \sigma$ содержит около $68\%$ вероятностной массы.
* От $a - 2\sigma$ до $a + 2\sigma$ — около $95\%$ массы (__правило двух сигм__).
* От $a - 3\sigma$ до $a + 3\sigma$ — около $99.7\%$ массы (__правило трёх сигм__).

Вероятностная масса - вероятность того, что данная величина попадает в интервал.

Для начала решим самые простые варианты:

* от 166 см до 182 см => от 174-8 до 174+8
по правилу одной сигмы, вероятность попасть в этот интервал равна **68%**

* от 158 см до 190 см => от 174-16 до 174+16
по правилу двух сигм, вероятность попасть в этот интервал равна **95%**

* не выше 150 см или не ниже 198 см => нужно **не попасть** от 174-24 до 174+24, по правилу трех сигм, 100% - 99.7%(вероятсность попасть в этот интервал) = **0.3%**

* больше 182

  182 - 174 = 8(одна сигма)

   По правилу одной сигмы, вероятность составляет 68%, но так как нам нужно попасть в интервал только > 174+8 (половина от   одной сигмы) получаем такую вероятсность:

  **(100 - 68)/2 = 16%**

* ниже 166

  174 - 166 = 8(одна сигма)

   Тут случай точно такой же, как и предыдущий

  **(100 - 68)/2 = 16%**

* больше 190

  190 - 174 = 16(две сигмы)

   По правилу двух сигм, вероятность составляет 95%, но так как нам нужно попасть в интервал только > 174+16(половина от интервала двух сигм), получаем такую вероятсность:

  **(100 - 95)/2 = 2.5%**

* не выше 150 см или не ниже 190 см

  посчитаем вероятности по отдельности:
  
  не выше 150(174-150=24), то есть **не попасть в интервал трех сигм** только с одной стороны => 0.3/2 = **0.15%**
  
  не ниже 190(190-174=16), то есть **не попасть в интервал двух сигм** только с одной стороны => 5/2 = **2.5%**
  
  и теперь сложим вероятности(так как в условии стоит **ИЛИ**) и получим: 0.15 + 2.5 = **2.65%**

* от 166 см до 190 см

  вероятность попасть в интервал от 158 до 190(две сигмы) равняется **95%**, вычтем из него половину интервала одной сигмы(потому что 174-166=8), и получим искомую вероятность:
  
  95 - (68/2) = **61%**
