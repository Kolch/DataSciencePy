```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
data = boston["data"]
feature_names = boston["feature_names"]
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
target = boston["target"]
y = pd.DataFrame(target, columns=["price"])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score?
```

```python
lr_r2_score = r2_score(y_test, y_pred)
lr_r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
model.fit(X_train, y_train.values[:,0])
```

```python
y_pred_ = model.predict(X_test)
```

```python
RF_r2_score = r2_score(y_test, y_pred_)
RF_r2_score
```

```python
print(lr_r2_score, RF_r2_score)
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
model = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
model.best_params_
```

```python
y_pred_proba = model.predict_proba(X_test)[:,1]
```

```python
y_pred_proba
```

```python
roc_auc_score(y_test, y_pred_proba)
```
