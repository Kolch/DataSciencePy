```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# y(k,x) = cos(kx)
```

```python
x = np.arange(0, 80)
y1 = np.cos(x)
y2 = np.cos(19*x)
```

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080', 'family': 'Calibri'}
label_dict = {'fontsize': '14', 'color': '#808080', 'family': 'Calibri'}
```

```python
plt.plot(x,y1, color='orange')
plt.plot(x,y2, color='red')

plt.title('Графики f(k,x)', fontdict=title_dict)
plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x,k)', fontdict=label_dict)

plt.legend(labels=['k=1', 'k=19'])
plt.show()
```
