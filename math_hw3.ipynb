```python
def vLength(vector):
    sqrt = 0
    for i in vector:
        sqrt = sqrt + i**2
    result = np.sqrt(sqrt)
    print(result)
```

```python
# сделал передачу массива чтобы было можно измерять векторы в любом измерении
vector = [10,10,0]
vLength(vector)
```

```python
def polarToCart(R, a):
    x = R* np.cos(a)
    y = R* np.sin(a)
    result = [x,y]
    return result
```

```python
x,y = polarToCart(12,45)
print(x)
print(y)
```

```python
def equations1(p):
    x,y = p
    return (x*2-y-1, np.exp(x) + x*(1-y)-1)
x1, y1 = fsolve(equations1, (-10,10))
print(x1,y1)
```

```python
def equations2(p):
    x,y = p
    return (x*2-y-1, np.exp(x) + x*(1-y)>1)
x2, y2 = fsolve(equations2, (-10,10))
print(x2,y2)
```
